```python
#interactive view of data
import nibabel as nib
import matplotlib.pyplot as plt
from ipywidgets import interact

img = nib.load('/content/3702_left_knee.nii')
input_data = img.get_fdata()

def view_slice(z):
    plt.imshow(input_data[:, :, z], cmap='gray')
    plt.title(f'Axial Slice {z}')
    plt.axis('off')
    plt.show()

interact(view_slice, z=(0, input_data.shape[2]-1))
```
